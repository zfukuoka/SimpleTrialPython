# ベルヌーイ試行の特定回数に対する確率のプロット

## ポアゾン分布の例

* 期待値 $$ λ = 2.0$$
* 分散 $$ \lambda $$
* 試行回数 $$ n = 16 $$
* 式 $$ P(k) = e^-λ \frac{λ^k}{k!} $$

```python
# ベルヌーイ試行の特定回数に対する確率のプロット

import matplotlib.pyplot as plt
import numpy as np
import math

ramda = 2.0
# x = np.linspace(0, 16, 17)
x = np.arange(0, 17)
factorials = np.array([math.factorial(i) for i in x])
y = np.exp(ramda * -1) * np.power(ramda, x) / factorials

# print(x.shape)
# print(y.shape)

plt.plot(x, y)
plt.xlabel("x number of success")
plt.ylabel("y rate")
plt.title("Plot Bernoulli trial")
plt.grid(True)
plt.show()
```
