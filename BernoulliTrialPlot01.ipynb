# ベルヌーイ試行の特定回数に対する確率のプロット

## ポアゾン分布の例

* 期待値 $$ λ = 2.0$$
* 分散 $$ \lambda $$
* 試行回数 $$ n = 16 $$
* 式 $$ P(k) = e^-λ \frac{λ^k}{k!} $$

```python
# ベルヌーイ試行の特定回数に対する確率のプロット

import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.special

numberOfTrial = 1200
rate = 1/1200
ramda = rate * numberOfTrial
# print(ramda)

# x = np.linspace(0, numberOfTrial, numberOfTrial+1)
x = np.arange(0, numberOfTrial+1)
# factorials = np.array([math.factorial(i) for i in x])
factorials = scipy.special.factorial(x)

y = np.exp(ramda * -1) * np.power(ramda, x) / factorials
print(y.dtype)

# print(x.shape)
# print(y.shape)

plt.plot(x, y)
plt.xlabel("x number of success")
plt.ylabel("y rate")
plt.xlim(0, 10)
plt.title("Plot Bernoulli trial")
plt.tick_params(axis='x', which='major', width=1, length=10)
plt.tick_params(axis='x', which='minor', width=1, length=1)
plt.grid(True)
plt.show()
```
