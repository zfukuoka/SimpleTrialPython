# ベルヌーイ試行の特定回数に対する確率のプロット

## ポアゾン分布の例

* 期待値 $$ λ = 1.0, 2.0, 4.0, 8.0 $$
* 分散 $$ \lambda $$
* 試行回数 $$ n = 1200 $$
* 式 $$ P(k) = e^-λ \frac{λ^k}{k!} $$

```python
# ベルヌーイ試行の特定回数に対する確率のプロット

import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.special

# 試行回数
numberOfTrial = 1200

# 試行の確率
rate = 1/1200
ramda = rate * numberOfTrial
rate_b = 1/600
ramda_b = rate_b * numberOfTrial
rate_c = 1/300
ramda_c = rate_c * numberOfTrial
rate_d = 1/150
ramda_d = rate_d * numberOfTrial
# print(ramda)

# 横軸の成功回数を試行回数分生成
# x = np.linspace(0, numberOfTrial, numberOfTrial+1)
x = np.arange(0, numberOfTrial+1)

# 予め階乗を算出したうえで、ポアゾン分布の式を当てはめ
# factorials = np.array([math.factorial(i) for i in x])
factorials = scipy.special.factorial(x)
y = np.exp(ramda * -1) * np.power(ramda, x) / factorials
y_b = np.exp(ramda_b * -1) * np.power(ramda_b, x) / factorials
y_c = np.exp(ramda_c * -1) * np.power(ramda_c, x) / factorials
y_d = np.exp(ramda_d * -1) * np.power(ramda_d, x) / factorials

# print(y.dtype)
# print(x.shape)
# print(y.shape)
print(y[0:5])
print(y_b[0:5])
print(y_c[0:5])
print(y_d[0:5])

# 横軸を成功回数、縦軸に割合でプロット
plt.plot(x, y, label="ramda = 1")
plt.plot(x, y_b, label="ramda = 2")
plt.plot(x, y_c, label="ramda = 4")
plt.plot(x, y_d, label="ramda = 8")
plt.legend()
plt.xlabel("x number of success")
plt.ylabel("y rate")
plt.xlim(0, 10)
plt.title("Plot Bernoulli trial")
plt.tick_params(axis='x', which='major', width=1, length=10)
plt.tick_params(axis='x', which='minor', width=1, length=1)
plt.grid(True)
plt.show()
```
