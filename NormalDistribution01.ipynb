# 正規分布のプロット

* 期待値 $$ μ $$
* 分散 $$ \sigma^2 $$
* 式 $$ f(x) =  \frac{1}{\sqrt{2πσ^2}} \exp{\biggl( - \frac{(x - μ)^2}{2σ^2} \biggr) }  $$

```python
# 正規分布のプロット

import matplotlib.pyplot as plt
import numpy as np


# 平均と標準偏差を定義
mu = 0
sigma = 1
sigma_b = 0.5
sigma_c = 1.5

# xの値を生成
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
x_b = np.linspace(mu - 3 * sigma_b, mu + 3 * sigma_b, 100)
x_c = np.linspace(mu - 3 * sigma_c, mu + 3 * sigma_c, 100)

# 正規分布の確率密度関数を計算
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
y_b = (1 / (sigma_b * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_b - mu) / sigma_b) ** 2)
y_c = (1 / (sigma_c * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_c - mu) / sigma_c) ** 2)

# +1/+2//-1/-2/ σの算出
x_sample = np.array([mu - 2 * sigma, mu - sigma, mu + sigma, mu + 2 * sigma])
y_sample = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_sample - mu) / sigma) ** 2)
x_sample_b = np.array([mu - 2 * sigma_b, mu - sigma_b, mu + sigma_b, mu + 2 * sigma_b])
y_sample_b = (1 / (sigma_b * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_sample_b - mu) / sigma_b) ** 2)
x_sample_c = np.array([mu - 2 * sigma_c, mu - sigma_c, mu + sigma_c, mu + 2 * sigma_c])
y_sample_c = (1 / (sigma_c * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_sample_c - mu) / sigma_c) ** 2)

# print(y.dtype)
# print(x.shape)
# print(y.shape)
print(y[0:5])
print(y_b[0:5])
print(y_c[0:5])

# 横軸を成功回数、縦軸に割合でプロット
plt.plot(x, y, label="sigma = 1.0")
plt.plot(x_b, y_b, label="sigma = 0.5")
plt.plot(x_c, y_c, label="sigma = 1.5")
plt.plot(x_sample, y_sample, 'o', label="sample")
plt.plot(x_sample_b, y_sample_b, 'o')
plt.plot(x_sample_c, y_sample_c, 'o')
plt.legend()
plt.xlabel("x")
plt.ylabel("Probability Density")
# plt.xlim(0, 10)
plt.title("Normal Distribution trial")
# plt.tick_params(axis='x', which='major', width=1, length=10)
# plt.tick_params(axis='x', which='minor', width=1, length=1)
# plt.grid(True)
plt.show()
```
