# 正規分布のプロット

* 期待値 $$ μ $$
* 分散 $$ \sigma^2 $$
* 式 $$ f(x) =  \frac{1}{\sqrt{2πσ^2}} \exp{\biggl( - \frac{(x - μ)^2}{2σ^2} \biggr) }  $$

```python
# ベルヌーイ試行の特定回数に対する確率のプロット

import matplotlib.pyplot as plt
import numpy as np


# 平均と標準偏差を定義
mu = 0
sigma = 1

# xの値を生成
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)

# 正規分布の確率密度関数を計算
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# print(y.dtype)
# print(x.shape)
# print(y.shape)
print(y[0:5])

# 横軸を成功回数、縦軸に割合でプロット
plt.plot(x, y, label="aa")
plt.legend()
plt.xlabel("x")
plt.ylabel("Probability Density")
# plt.xlim(0, 10)
plt.title("Plot Bernoulli trial")
# plt.tick_params(axis='x', which='major', width=1, length=10)
# plt.tick_params(axis='x', which='minor', width=1, length=1)
# plt.grid(True)
plt.show()
```
